```python
from pathlib import Path
import pandas as pd

def load_csv(file_name = None, dir = "Data_csv_0539"):   
    link_path = dir
    q = Path(link_path)
    while not q.exists():
        link_path = "../" + link_path
        q = Path(link_path)
        if len(link_path) > 500: raise Exception(f"Directory {dir} not found.") 
    if not file_name: return list(q.glob('*.csv'))
    f = q / file_name
    if not f.exists():
        raise Exception("File not found.")    
    file_csv = f.resolve()
    return pd.read_csv(file_csv, low_memory=False)

from pathlib import Path
import sys

def load_pexplorer():   
    link_path = "pexplorer"
    q = Path(link_path)
    while not q.exists():
        link_path = "../" + link_path
        q = Path(link_path)
        if len(link_path) > 500: raise Exception("Directory pexplorer not found.") 
    sys.path.append(link_path)
    import pexplorer as px
    return px

px = load_pexplorer()
```
